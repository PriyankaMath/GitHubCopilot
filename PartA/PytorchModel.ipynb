```python
# Build a simple neural network using Pytorch

import torch
import torch.nn as nn

#define variables like inputsize, hidden size, output size, batch size and learning rate
n_input, n_hidden, n_output, batch_size, learning_rate = 10, 15, 1, 100, 0.01

#Randomly initialize dummy input and output data
data_x = torch.randn(batch_size, n_input)
data_y = (torch.rand(size=(batch_size, n_output)) < 0.5).float()
```

```python
print(data_x.size())
print(data_y.size())
```

```python
# Create a sequential neural network with output sigmoid layer

model = nn.Sequential(nn.Linear(n_input, n_hidden), nn.ReLU(), nn.Linear(n_hidden, n_output), nn.Sigmoid())
print(model)
```

```python
# Define the loss function and optimizer for gradient descent

loss_function = nn.MSELoss();
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# Define the training loop with forward, backward pass and gradient descent
losses = []
for epoch in range(5000):
    # Forward pass
    y_pred = model(data_x)

    # Compute loss
    loss = loss_function(y_pred, data_y)
    losses.append(loss.item())

    model.zero_grad()

    # Backward pass
    loss.backward()

    # Update weights
    optimizer.step()
```

```python
# Plot the loss function
import matplotlib.pyplot as plt
plt.plot(losses)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.title('Learning rate = ' + str(learning_rate))
plt.show()
```
