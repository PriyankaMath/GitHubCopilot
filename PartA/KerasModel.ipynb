```python
# Import libraries to build a tensorflow model
import tensorflow as tf
import numpy as np

# function that generates an XOR dataset
def generate_xor_dataset():
    x_data = np.array([[0,0],[0,1],[1,0],[1,1]])
    y_data = np.array([[0],[1],[1],[0]])
    return x_data, y_data

# function that creates a sequential model for the XOR dataset
# this model will have two hidden layers that uses relu as activation function
def create_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(2, input_shape=(2,), activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

#function to train the model, takes epoch as input
def train_model(model, epochs):
    x_data, y_data = generate_xor_dataset()
    model.fit(x_data, y_data, epochs=epochs)
    return model

#function to test the model and return the accuracy
def test_model(model):
    x_data, y_data = generate_xor_dataset()
    return model.evaluate(x_data, y_data)

#create the model
model = create_model()
#train the model
model = train_model(model, 100)
#test the model
print("Accuracy: ",test_model(model))
```
